## Read `metadata` table

```python
metadata = pd.read_table('combined_tables/concatenated_metadata_all.txt', sep='\s+')
print(metadata.shape)
metadata.head()
```

This looks like `metadata` is the output from the STAR aligner

```python
metadata_t = metadata.T
print(metadata_t.shape)
metadata_t.head()
```

Remove all whitespace (e.g. `\t`) characters and percentage signs

```python
metadata_t_stripped = metadata_t.applymap(lambda x: x.strip().strip('%') if isinstance(x, str) else x)
print(metadata_t_stripped.shape)
metadata_t_stripped.head()
```

```python
metadata_t_stripped.dtypes
```

Convert strings to numbers

```python
pd.options.display.max_columns = 50
```

```python
metadata_t_numeric = metadata_t_stripped.apply(pd.to_numeric, errors='ignore')
print(metadata_t_numeric.shape)
print(metadata_t_numeric.dtypes)
metadata_t_numeric.head()
```

```python
metadata_t_dropna = metadata_t_numeric.dropna(how='all', axis=1)
print(metadata_t_dropna.shape)
metadata_t_dropna.head()
```

```python
metadata_t_dropna.columns
```

argh columns have spaces in them .. time to strip them again

```python
metadata_t_dropna.columns = metadata_t_dropna.columns.map(lambda x: x.strip())
metadata_t_dropna.columns
```

```python
sns.boxplot(metadata_t_dropna['Uniquely mapped reads %'])
```

```python
unmapped_cols = [x for x in metadata_t_dropna.columns if x.startswith('% of reads')]
unmapped_cols
```

```python
sns.boxplot(metadata_t_dropna[unmapped_cols], orient='h')
```

All those cells around 40% mapped to multiple loci are probably interesting cells from a tissue that has a lot of homologous gene expression

This is the output from HTSeq / featurecounts

```python
all_zeros = (expression == 0).all(axis=1)
all_zeros.sum()
```

```python
expression_nonzero = expression.loc[~all_zeros]
print(expression_nonzero.shape)
expression_nonzero.head()
```

```python
expression_nonzero.tail()
```

```python
all_zeros = (expression == 0).all(axis=0)
all_zeros.sum()
```

So there aren't any cells that have zero expression

```python
from sklearn.decomposition import PCA
```

```python
reducer = PCA()
reduced = pd.DataFrame(reducer.fit_transform(expression_nonzero.T), index=expression_nonzero.columns)
print(reduced.shape)
reduced.head()
```

```python
reduced_metadata = reduced.join(metadata_tissues)
print(reduced_metadata.shape)
reduced_metadata.head()
```

```python
reduced_subset = reduced_metadata[list(range(10)) + ['tissue', 'subtissue', 'tissue_subtissue', 
                                                     'Number of input reads', 'Uniquely mapped reads number']]
print(reduced_subset.shape)
reduced_subset.head()
```
