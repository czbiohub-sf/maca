```python
import pandas as pd
csv = '/Users/olgabot/Downloads/maca_3month_original_vs_cleaned_plates.csv'
df = pd.read_csv(csv, encoding="ISO-8859-1")
print(df.shape)
cols_to_drop = [x for x in df if x.startswith('Unnamed')]
df = df.drop(cols_to_drop, axis=1)
df.head()
```

```python
replacer = {
    'http://purl.obolibrary.org/obo/OMIT_0023178': 
        'http://purl.obolibrary.org/obo/CL_0000115'}
df['Term IRI'] = df['Term IRI'].str.replace('http://purl.obolibrary.org/obo/OMIT_0023178',
                                            'http://purl.obolibrary.org/obo/CL_0000115')
```

```python
df['ontology_id'] = df['Term IRI'].map(
    lambda x: x.split('/')[-1].replace('_', ':') 
    if isinstance(x, str) else x)
df.head()
```

```python
for ontology_id in df['ontology_id'].unique():
    if not pd.isnull(ontology_id):
        term = cell_ontology[ontology_id]
        print('\n\n--- term:', term, '---')
        print('parents:')
        for parent in term.parents:
            print(f'\t{parent}')
        print('\n\tchildren:')
        for child in term.children:
            print(f'\t\t{child}')
            
        print('\n\tSynonyms:')
        for synonym in term.synonyms:
            print(f'\t\t{synonym}')
```

```python
ontology_df = pd.Series({term.id: term.name for term in cell_ontology}).to_frame()
print(ontology_df.shape)
ontology_df.index.name = 'cell_ontology_id'
ontology_df = ontology_df.rename(columns={0: 'Class'})
ontology_df['Term IRI'] = 'http://purl.obolibrary.org/obo/' + ontology_df.index.str.replace(':', '_')
ontology_df.head()
```

```python
from itertools import chain
```

```python
synonym_names = list(chain.from_iterable([s.desc for s in term.synonyms] for term in cell_ontology))
synonym_ids = list(chain.from_iterable([[term.id]*len(term.synonyms) 
                      for term in cell_ontology]))
```

```python
synonym_names
```

```python
synonym_df = pd.Series(synonym_names, index=synonym_ids, name='Synonym').to_frame()
synonym_df.index.name = 'cell_ontology_id'
synonym_df['Term IRI'] = 'http://purl.obolibrary.org/obo/' + synonym_df.index.str.replace(':', '_')

print(synonym_df.shape)
synonym_df.head()
```
