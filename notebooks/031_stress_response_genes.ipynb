```python
import pandas as pd
import os
```

```python
mgi_ids = pd.read_table('/Users/olgabot/Downloads/mouse_genome_informatics_ids.txt', index_col=0)
print(mgi_ids.shape)
mgi_ids.head()
```

```python
stress_response = pd.read_table('/Users/olgabot/Downloads/cellular_stress_response_mouse.txt', header=None)
print(stress_response.shape)
stress_response.head()
```

```python
stress_response['mgi_id'] = 'MGI:' + stress_response[0].str.split(':').str[-1]
stress_response.head()
```

```python
stress_response_unique = stress_response.drop_duplicates(subset=['mgi_id'])
print(stress_response_unique.shape)
stress_response_unique.head()
```

```python
stress_response_unique_symbols = stress_response_unique.join(mgi_ids, on='mgi_id')
print(stress_response_unique_symbols.shape)
stress_response_unique_symbols.head()
```

```python

stress_response_unique_symbols['Marker Symbol'].to_csv('/Users/olgabot/code/maca/data/cellular_response_to_stress_genes_go-0033554.csv')
```
